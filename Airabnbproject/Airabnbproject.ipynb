```python
import pandas as pd
import random

sample_amenities = [
    ['WiFi', 'Kitchen'], ['TV', 'Internet'], ['Heating', 'Air conditioning'],
    ['WiFi', 'TV'], ['Kitchen', 'Washer'], ['Internet', 'Heating'],
    ['Air conditioning', 'Kitchen'], ['WiFi', 'Washer'], ['TV', 'Heating'],
    ['Internet', 'WiFi'], ['Kitchen', 'Heating'], ['TV', 'Air conditioning'],
    ['Washer', 'WiFi'], ['Internet', 'Kitchen'], ['Heating', 'WiFi']
] * (931 // 15) + [['WiFi', 'Kitchen']] * (931 % 15)

sample_availability = [
    {'availability_30': 15}, {'availability_30': 20}, {'availability_30': 5},
    {'availability_30': 12}, {'availability_30': 8}, {'availability_30': 22},
    {'availability_30': 14}, {'availability_30': 18}, {'availability_30': 25},
    {'availability_30': 10}, {'availability_30': 7}, {'availability_30': 16},
    {'availability_30': 9}, {'availability_30': 11}, {'availability_30': 6}
] * (931 // 15) + [{'availability_30': 15}] * (931 % 15)

sample_review_scores = [
    {'rating': 4.5}, {'rating': 4.8}, {'rating': 4.7},
    {'rating': 4.3}, {'rating': 4.1}, {'rating': 4.9},
    {'rating': 4.4}, {'rating': 4.6}, {'rating': 4.2},
    {'rating': 4.0}, {'rating': 3.9}, {'rating': 4.8},
    {'rating': 4.7}, {'rating': 4.5}, {'rating': 4.6}
] * (931 // 15) + [{'rating': 4.5}] * (931 % 15)


df1['amenities'] = sample_amenities
df1['availability'] = sample_availability
df1['review_scores'] = sample_review_scores
df1['amenities_count'] = df1['amenities'].apply(lambda x: len(x) if isinstance(x, list) else 0)

df1['availability_30'] = df1['availability'].apply(lambda x: x.get('availability_30') if isinstance(x, dict) else None)

df1['review_score_rating'] = df1['review_scores'].apply(lambda x: x.get('rating') if isinstance(x, dict) else None)
df1 = df1.drop(columns=['amenities', 'availability', 'review_scores'])
print(df1.head())
```
